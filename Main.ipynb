```python
import json
import time
```

```python
def loadJson():
    sum=0
    while(True):
        file=open("records.json","r")
        text=file.read()
        file.close()
        record=json.loads(text)
        bill=display(record)
        sum=sum+bill
        exit_status=input("Press 'Y' to Continue Shopping / Press any other Key to Exit:")
        if(exit_status=='Y' or exit_status=="y"):
            continue
        else:
            print()
            print("THANKS FOR SHOPPING AT VIKAS MART")
            print("Bill="+str(sum))
            break
```

```python
def display(record):
    print()
    for i in record.keys():
        print(record[i]["name"]+" -->Rs"+str(record[i]["price"])+" per "+record[i]["quantity"])
    print("-----------------------------------------------------------------------------")
    bill=getDetails(record)
    return bill
```

```python
def getDetails(record):
    print()
    choice=input("What You Want To Buy:")
    stock=int(input("Enter The Quantity Of The "+ choice +" You Want To Buy:"))
    bill=process(choice,stock,record)
    return bill
```

```python
def process(choice, stock,record):
    flag=False
    bill=0
    for i in record.keys():
        if(choice==record[i]["name"]):
            flag=True
            if(stock<=record[i]["stock"]):
                record[i]["stock"]=record[i]["stock"]-stock
                updateRecord(record)
                bill=billing(choice,stock,record)
                break
            else:
                print("Product is Out of Stock")
                break
    if(flag==False):
        print("This Product is not Sold by us")
    return bill
                
            
```

```python
def updateRecord(record):
    js=json.dumps(record)
    file=open("records.json","w")
    file.write(js)
    file.close()
```

```python
def billing(choice,stock,record):
    for i in record.keys():
        if(choice==record[i]["name"]):
            bill=stock*record[i]["price"]
            st={time.ctime():{choice:stock}}
            js=json.dumps(st)
            file=open("Sales.json","a")
            file.write(js)
            file.write("\n")
            file.close()
            return bill
```
